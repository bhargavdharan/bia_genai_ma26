Run inference with two different Mistral variants:   - `Mistral-7B-Instruct-v0.3`   - `Mistral-Small-3.1-24B-Instruct-2503`The `device=0` argument tells the pipeline to use your GPU. If you do not have a GPU, change it to `device=-1` to run on CPU (slower).

## Section 2: Load the Hugging Face TokenSome models on Hugging Face require accepting a license or logging in. We store the token in a `.env` file and read it here. You can also pass `token=hf_token` to the `pipeline()` call later.

```python
from dotenv import load_dotenv
import os

load_dotenv()

hf_token = os.getenv("HF_TOKEN")
print(hf_token)
```

What this cell does:- Loads environment variables from `.env`.- Reads `HF_TOKEN` into `hf_token`.- Prints the token (partially shown) so you can confirm it was loaded.**Tip:** In real projects, avoid printing secrets. It is shown here for learning purposes.

## Section 3: Run Mistral-7B-Instruct`Mistral-7B-Instruct-v0.3` is a 7-billion-parameter instruction-tuned model. It expects a chat-style prompt with roles: `user` and `assistant`.

```python
from transformers import pipeline
import torch


messages = [
    {"role": "user", "content": "What is your favourite condiment?"},
    {"role": "assistant", "content": "Well, I'm quite partial to a good squeeze of fresh lemon juice. It adds just the right amount of zesty flavour to whatever I'm cooking up in the kitchen!"},
    {"role": "user", "content": "Do you have mayonnaise recipes?"}
]

chatbot = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.3", device=0)
chatbot(messages)
```

Line-by-line explanation:- `from transformers import pipeline` imports the high-level Hugging Face inference helper.- `import torch` brings in the PyTorch backend.- `messages` is a list of dictionaries representing the conversation history. The model sees the prior assistant reply and the new user question.- `pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.3", device=0)` loads the model onto GPU 0 and creates a text-generation pipeline.- `chatbot(messages)` runs inference and returns the model's continuation.Expected output: a generated assistant response to "Do you have mayonnaise recipes?", possibly continuing the condiment conversation.

## Section 4: Run Mistral-Small-3.1-24B-InstructThis is a larger 24-billion-parameter Mistral model. The code is identical to the previous section; only the model name changes. Larger models often produce more detailed or coherent answers but require more GPU memory and time.

```python
from transformers import pipeline
import torch


messages = [
    {"role": "user", "content": "What is your favourite condiment?"},
    {"role": "assistant", "content": "Well, I'm quite partial to a good squeeze of fresh lemon juice. It adds just the right amount of zesty flavour to whatever I'm cooking up in the kitchen!"},
    {"role": "user", "content": "Do you have mayonnaise recipes?"}
]

chatbot = pipeline("text-generation", model="mistralai/Mistral-Small-3.1-24B-Instruct-2503", device=0)
chatbot(messages)
```

What changes here:- The model ID is now `mistralai/Mistral-Small-3.1-24B-Instruct-2503`.- The prompt (`messages`) is the same, so you can directly compare outputs between the 7B and 24B versions.If your GPU does not have enough VRAM, this cell may fail with an out-of-memory error. In that case, use `device=-1` for CPU inference or a smaller model.
